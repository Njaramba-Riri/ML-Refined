#### Resampling, Shifting, and Windowing

```python
from pandas_datareader import data

goog=data.DataReader('GOOG', start='2018', end='2021', data_source='yahoo')
goog.head()
```

```python
goog.head()
```

```python
goog=goog['Close']
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
```

```python
goog.plot()
plt.title("Yahoo's closing stock over time")
```

##### Resampling and converting frequencies

```python
goog.plot(alpha=0.5, style='-')
goog.resample('BA').mean().plot(style=':')
goog.asfreq('BA').plot(style='--');
plt.legend(['input', 'resample', 'asfreq'],
          loc='upper left')
plt.title("Resampling Yahoo's stock price ")
```

```python
fig, ax=plt.subplots(2, sharex=True)
data=goog.iloc[:10]

data.asfreq('D').plot(ax=ax[0], marker='o')

data.asfreq('D', method='bfill').plot(ax=ax[1], style='-o')
data.asfreq('D', method='ffill').plot(ax=ax[1], style='--o')
ax[1].legend(["back-fill", "foward-fill"])
plt.title('Comparison between forward-fill and back-fill interpolation');
```

###### Time-shifts

```python
fig, ax=plt.subplots(3, sharey=True)

#apply Frequency to the data
goog=goog.asfreq('D', method='pad')

goog.plot(ax=ax[0])
goog.shift(900).plot(ax=ax[1])
goog.tshift(900).plot(ax=ax[2])

#legends and annotations
local_max=pd.to_datetime('2019-03-16')
offset=pd.Timedelta(900,'D')

ax[0].legend(['input'], loc=2)
ax[0].get_xticklabels()[4].set(weight='heavy', color='red')
ax[0].axvline(local_max, alpha=0.3, color='red')

ax[1].legend(['shift(900)'], loc=2)
ax[1].get_xticklabels()[4].set(weight='heavy', color='red')
ax[1].axvline(local_max + offset, alpha=0.3, color='red')

ax[2].legend(['tshift(900)'],loc=2)
ax[2].get_xticklabels()[1].set(weight='heavy', color='red')
ax[2].axvline(local_max + offset, alpha=0.3, color='red');
```

```python
ROI=100*(goog.shift(-365)/goog-1)
ROI.plot()
plt.ylabel('% Return on Investment');
```

###### Rolling windows

```python
rolling=goog.rolling(365, center=True)

data=pd.DataFrame({'input':goog,
                  'one-year rolling_mean':rolling.mean(),
                  'one-year rolling_std':rolling.std()})
ax=data.plot(style=['-','--',':'])
ax.lines[0].set_alpha(0.3)
plt.title("Rolling statistics on Yahoo's stock");
```

### High-performance Pandas: eval() and query()

```python
import numpy as np
```

```python
rng=np.random.RandomState(42)
x=rng.rand(16)
y=rng.rand(16)
%timeit x+y
```

```python
%timeit np.fromiter((xi+yi for xi, yi in zip(x,y)),dtype=x.dtype, count=len(x))
```

```python
mask=(x>0.5)&(y<0.5)
```

```python
tmp1=(x>0.5)
tmp2=(y<0.5)
mask=tmp1&tmp2
```

```python
import numexpr
mask_numexpr=numexpr.evaluate('(x>0.5) &(y<0.5)')
np.allclose(mask, mask_numexpr)
```

### Pandas .eval() for efficient Operations

```python
nrows, ncols=100000, 100
rng=np.random.RandomState(42)
df1, df2, df3, df4=(pd.DataFrame(rng.rand(nrows, ncols))
                   for i in range(4))
```

```python
%timeit df1+df2+df3+df4
```

```python
%timeit pd.eval('df1+df2+df3+df4')
```

```python
np.allclose(df1+df2+df3+df4,
           pd.eval('df1+df2+df3+df4'))
```

### DataFrame.eval() for Column-Wise Operations

```python
df=100*pd.DataFrame(rng.rand(1000,3),columns =['A','B', 'C'])
df.head()
```

```python
result1 = (df['A'] + df['B']) / (df['C'] - 1)
result2 = pd.eval("(df.A + df.B) / (df.C - 1)")
np.allclose(result1, result2)
```

```python
result3=df.eval('(A+B)/(C-1)')
np.allclose(result2,result3)
```

```python
df.head()
```

```python
df.eval('D=(A+B)/C', inplace=True)
df.head()
```

```python
df.eval('D=(A-B)/C', inplace=True)
df.head()
```
