```python
#pivot table syntax
titanic.pivot_table('Respondent Age', columns='country', index='gender_of_respondent')
```

```python
#Birthrate Data
uspop=pd.read_csv('~/downloads/student-mat.csv', sep=';')
uspop.head()
```

```python
uspop.pivot_table('traveltime', index='school', columns='sex',aggfunc='sum')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
sns.set()#seaborn style
uspop.pivot_table(['G3','G2','G1'], index='traveltime').plot()
plt.title("Score in Maths vs Traveltime to school")
plt.ylabel('score in Mathematics');
```

```python
#Further data exploration
quartiles=np.percentile(uspop['absences'],[25,50,75])
mu=quartiles[1]
sig=0.74*(quartiles[2]-quartiles[0])
```

```python
uspop=uspop.query('(absences>@mu-5*@sig)&(absences<@mu+5*@sig)')
```
