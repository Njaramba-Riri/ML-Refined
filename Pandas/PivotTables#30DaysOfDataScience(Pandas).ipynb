```python
brecord=pd.read_csv('births.csv')
brecord.head()
```

```python
brecord['decade'] =10*(brecord['year']//10)
brecord.pivot_table('births', index='decade', columns='gender', aggfunc='sum')
```

```python
brecord.head()
```

```python

brecord.pivot_table('births', index='year', columns='gender', aggfunc='sum').plot()
plt.ylabel('Births per year')
```

```python
#Further data exploration
quartiles=np.percentile(brecord['births'],[25,50,75])
mu=quartiles[1]
sig=0.74*(quartiles[2]-quartiles[0])
```

```python
brecord =brecord.query('(births > @mu-5 * @sig) & (births < @mu + 5 * @sig)')
```

```python
brecord['day']=brecord['day'].astype(dtype=int)
```

```python
#create a datetime index from year,month,day
brecord.index=pd.to_datetime(10000*brecord.year+
                            100*brecord.month+
                            brecord.day, format='%Y-%m-%d')
brecord['dayofweek']=brecord.index.dayofweek
brecord
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
brecord.pivot_table('births',index='dayofweek',columns='decade',
                   aggfunc='mean').plot()
plt.gca().set_xticklabels(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])
plt.ylabel('mean births by day');
```
